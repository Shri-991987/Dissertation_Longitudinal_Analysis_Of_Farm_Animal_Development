### Nadaraya-Watson

```python
# Nadaraya–Watson Estimator on whole lambs dataset
from scipy.stats import norm

def gaussian_kernel(distance, bandwidth):
    return norm.pdf(distance / bandwidth)

def nadaraya_watson(x, X, Y, bandwidth):
    weights = gaussian_kernel(x - X, bandwidth)
    return np.sum(weights * Y) / np.sum(weights)

def estimate_weight(lamb_no, target_age, bandwidth):
    lamb_data = combined_df[combined_df['No.'] == lamb_no].dropna(subset=['Weight'])
    X = lamb_data['Age (Days)'].values
    Y = lamb_data['Weight'].values
    estimated_weight = nadaraya_watson(target_age, X, Y, bandwidth)
    return estimated_weight

# Example: Estimating the weight of lamb 611 at age 200 days
lamb_no = 611
target_age = 200
bandwidth = 10  # Bandwidth Adjustment

estimated_weight = estimate_weight(lamb_no, target_age, bandwidth)
print(f"Estimated weight of lamb {lamb_no} at age {target_age} days: {estimated_weight}")
```

### LOO-CV - Bandwidth Estimation

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import mean_squared_error
import numpy as np

def cross_validation_bandwidth_loocv(lamb_no, bandwidths):
    lamb_data = combined_df[combined_df['No.'] == lamb_no].dropna(subset=['Weight'])
    X = lamb_data['Age (Days)'].values
    Y = lamb_data['Weight'].values

    loo = LeaveOneOut()
    avg_mse = []

    for bandwidth in bandwidths:
        mse_list = []

        for train_index, val_index in loo.split(X):
            X_train, X_val = X[train_index], X[val_index]
            Y_train, Y_val = Y[train_index], Y[val_index]

            Y_pred = nadaraya_watson(X_val[0], X_train, Y_train, bandwidth)
            mse = mean_squared_error(Y_val, [Y_pred])
            mse_list.append(mse)

        avg_mse.append(np.mean(mse_list))
    
    best_bandwidth = bandwidths[np.argmin(avg_mse)]
    return best_bandwidth, avg_mse

# Define a range of bandwidths to test
bandwidths = np.linspace(5, 50, 10)

# Find the best bandwidth for lamb 611
best_bandwidth, avg_mse = cross_validation_bandwidth_loocv(lamb_no=611, bandwidths=bandwidths)

print(f"Best bandwidth: {best_bandwidth}")
```

```python
import numpy as np

def weight_function(age, target_age, c):
    return np.exp(-c * (age - target_age)**2)

def nadaraya_watson_normalized(target_age, X, Y, c):
    # Calculate weights using the weight function
    weights = weight_function(X, target_age, c)
    # Normalize the weights so that they sum to 1
    normalized_weights = weights / np.sum(weights)
    
    # Calculate the weighted average of Y
    estimated_weight = np.sum(normalized_weights * Y)
    
    return estimated_weight

def estimate_weight_nadaraya_watson(lamb_no, target_age, c):
    lamb_data = final_df2l[final_df2l['No.'] == lamb_no].dropna(subset=['Weight'])
    X = lamb_data['Age (Days)'].values
    Y = lamb_data['Weight'].values
    estimated_weight = nadaraya_watson_normalized(target_age, X, Y, c)
    return estimated_weight

# Example: Estimate the weight of lamb 611 at age 200 days
lamb_no = 612
target_age = 200
c = 0.001  # You can adjust the value of c

estimated_weight = estimate_weight_nadaraya_watson(lamb_no, target_age, c)
print(f"Estimated weight of lamb {lamb_no} at age {target_age} days: {estimated_weight}")
```
