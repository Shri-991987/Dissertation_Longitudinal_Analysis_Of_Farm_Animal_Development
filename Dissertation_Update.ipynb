### Nadaraya-Watson

```python
# Nadaraya–Watson Estimator on whole lambs dataset
from scipy.stats import norm

def gaussian_kernel(distance, bandwidth):
    return norm.pdf(distance / bandwidth)

def nadaraya_watson(x, X, Y, bandwidth):
    weights = gaussian_kernel(x - X, bandwidth)
    return np.sum(weights * Y) / np.sum(weights)

def estimate_weight(lamb_no, target_age, bandwidth):
    lamb_data = final_df2[final_df2['No.'] == lamb_no].dropna(subset=['Weight'])
    X = lamb_data['Age (Days)'].values
    Y = lamb_data['Weight'].values
    estimated_weight = nadaraya_watson(target_age, X, Y, bandwidth)
    return estimated_weight

# Example: Estimating the weight of lamb 611 at age 200 days
lamb_no = 611
target_age = 200
bandwidth = 30  # Bandwidth Adjustment

estimated_weight = estimate_weight(lamb_no, target_age, bandwidth)
print(f"Estimated weight of lamb {lamb_no} at age {target_age} days: {estimated_weight}")
```

```python
import numpy as np

def weight_function(age, target_age, c):
    return np.exp(-c * (age - target_age)**2)

def nadaraya_watson_normalized(target_age, X, Y, c):
    # Calculate weights using the weight function
    weights = weight_function(X, target_age, c)
    # Normalize the weights so that they sum to 1
    normalized_weights = weights / np.sum(weights)
    
    # Calculate the weighted average of Y
    estimated_weight = np.sum(normalized_weights * Y)
    
    return estimated_weight

def estimate_weight_nadaraya_watson(lamb_no, target_age, c):
    lamb_data = final_df2l[final_df2l['No.'] == lamb_no].dropna(subset=['Weight'])
    X = lamb_data['Age (Days)'].values
    Y = lamb_data['Weight'].values
    estimated_weight = nadaraya_watson_normalized(target_age, X, Y, c)
    return estimated_weight

# Example: Estimate the weight of lamb 611 at age 200 days
lamb_no = 612
target_age = 200
c = 0.001  # You can adjust the value of c

estimated_weight = estimate_weight_nadaraya_watson(lamb_no, target_age, c)
print(f"Estimated weight of lamb {lamb_no} at age {target_age} days: {estimated_weight}")
```

```python
# Local Linear Fit:

from sklearn.linear_model import LinearRegression

def weight_function(age, target_age, c):
    return np.exp(-c * (age - target_age)**2)


def weighted_local_linear_fit(target_age, X, Y, c):
    weights = weight_function(X, target_age, c)
    
    # Fit a weighted linear regression model
    model = LinearRegression()
    X_reshaped = X.reshape(-1, 1)
    model.fit(X_reshaped, Y, sample_weight=weights)
    
    # Predict the weight at the target age
    predicted_weight = model.predict([[target_age]])
    return predicted_weight[0]
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to compute Gaussian kernel
def gaussian_kernel(distance, bandwidth):
    return norm.pdf(distance / bandwidth)

# Function to compute Nadaraya-Watson estimate
def nadaraya_watson(x, X, Y, bandwidth):
    weights = gaussian_kernel(x - X, bandwidth)
    return np.sum(weights * Y) / np.sum(weights)

# Function to estimate weight for a given lamb and target age
def estimate_weight(lamb_no, target_age, bandwidth):
    lamb_data = final_df2[final_df2['No.'] == lamb_no].dropna(subset=['Weight'])
    X = lamb_data['Age (Days)'].values
    Y = lamb_data['Weight'].values
    estimated_weight = nadaraya_watson(target_age, X, Y, bandwidth)
    return estimated_weight

# Estimating the weight curve for lamb
def estimate_weight_curve(lamb_no, X, bandwidth):
    estimated_weights = []
    for age in X:
        estimated_weight = estimate_weight(lamb_no, age, bandwidth)
        estimated_weights.append(estimated_weight)
    return np.array(estimated_weights)

# Define the lamb number and the age range for estimation
lamb_no = 612
X = np.arange(0, 450, 1)  # Estimating weights from day 0 to 300
bandwidth = 30  # Bandwidth Adjustment

# Estimate the weight curve
estimated_weights = estimate_weight_curve(lamb_no, X, bandwidth)

# Plot the actual data and the estimated curve
plt.figure(figsize=(12, 6))
lamb_data = final_df2[final_df2['No.'] == lamb_no]
plt.plot(lamb_data['Age (Days)'], lamb_data['Weight'], 'o', label=f'Actual Data (Lamb {lamb_no})')
plt.plot(X, estimated_weights, label='Estimated Curve', color='red', linestyle='--')

plt.title('Estimated Growth Curve of Lamb 612')
plt.xlabel('Age (Days)')
plt.ylabel('Weight')
plt.legend()
plt.grid(True)
plt.show()
```

#### Trying the Estimated Curev Using OLS Local Linear Fit

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to compute Gaussian kernel
def gaussian_kernel(distance, bandwidth):
    return norm.pdf(distance / bandwidth)

# Function to compute local linear regression estimate
def local_linear_regression(x, X, Y, bandwidth):
    n = len(X)
    weights = gaussian_kernel(X - x, bandwidth)
    
    # Construct weight matrix
    W = np.diag(weights)
    
    # Construct design matrix
    X_mat = np.vstack((np.ones(n), X - x)).T
    
    # Compute (X^T W X)^-1 X^T W Y
    beta = np.linalg.inv(X_mat.T @ W @ X_mat) @ (X_mat.T @ W @ Y)
    
    # Predicted value at x
    y_pred = beta[0]
    
    return y_pred

# Function to estimate weight for a given lamb and target age using local linear regression
def estimate_weight_llr(lamb_no, target_age, bandwidth):
    lamb_data = final_df2[final_df2['No.'] == lamb_no].dropna(subset=['Weight'])
    X = lamb_data['Age (Days)'].values
    Y = lamb_data['Weight'].values
    estimated_weight = local_linear_regression(target_age, X, Y, bandwidth)
    return estimated_weight

# Estimating the weight curve for lamb using local linear regression
def estimate_weight_curve_llr(lamb_no, X, bandwidth):
    estimated_weights = []
    for age in X:
        estimated_weight = estimate_weight_llr(lamb_no, age, bandwidth)
        estimated_weights.append(estimated_weight)
    return np.array(estimated_weights)

# Define the lamb number and the age range for estimation
lamb_no = 612
X = np.arange(0, 430, 1)  # Estimating weights from day 0 to 300
bandwidth = 20  # Bandwidth Adjustment

# Estimate the weight curve using local linear regression
estimated_weights_llr = estimate_weight_curve_llr(lamb_no, X, bandwidth)

# Plot the actual data and the estimated curve
plt.figure(figsize=(12, 6))
lamb_data = final_df2[final_df2['No.'] == lamb_no]
plt.plot(lamb_data['Age (Days)'], lamb_data['Weight'], 'o', label=f'Actual Data (Lamb {lamb_no})')
plt.plot(X, estimated_weights_llr, label='Estimated Curve (Local Linear Regression)', color='red', linestyle='--')

plt.title('Estimated Growth Curve of Lamb')
plt.xlabel('Age (Days)')
plt.ylabel('Weight')
plt.legend()
plt.grid(True)
plt.show()
```

### Applying Multivariate Regression Model

```python
mul_df = final_df2.drop(['Type'],axis=1)
mul_df
```

```python
final_mul = pd.get_dummies(mul_df, columns=['Gender'], drop_first=True)
final_mul.dropna(subset=['Weight'], inplace=True)
final_mul.reset_index(drop=True, inplace=True)
```

```python
final_mul
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

# Define the independent variables (features) and dependent variable (target)
X = final_mul[['Age (Days)', 'Litter_size', 'Gender_M']]
y = final_mul['Weight']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Polynomial transformation (3rd order)
poly = PolynomialFeatures(degree=3)
X_poly_train = poly.fit_transform(X_train)
X_poly_test = poly.transform(X_test)

# Initialize and fit the model
model = LinearRegression()
model.fit(X_poly_train, y_train)

# Predict the weights for the test set
y_pred = model.predict(X_poly_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
coefficients = pd.DataFrame(model.coef_, X.columns, columns=['Coefficient'])
print(coefficients)
```

### Prediction and Plotting

```python
def predict_weights(lamb_no, age_range, litter_size, gender):
    X_pred = np.array([[age, litter_size, gender] for age in age_range])
    X_poly_pred = poly.transform(X_pred)
    predicted_weights = model.predict(X_poly_pred)
    return predicted_weights

```

```python
def predict_weights_for_lamb(df,lamb_no, age_range):
    # Extract the lamb's data for litter size and gender
    lamb_data = df[df['No.'] == lamb_no].iloc[0]
    litter_size = lamb_data['Litter_size']
    gender = lamb_data['Gender_M']
    
    # Prepare the input data for prediction
    X_pred = np.array([[age, litter_size, gender] for age in age_range])
    X_poly_pred = poly.transform(X_pred)
    predicted_weights = model.predict(X_poly_pred)
    
    return predicted_weights

# Function to plot the growth curve
def plot_growth_curve(df,lamb_no, age_range):
    # Predict weights for the lamb
    predicted_weights = predict_weights_for_lamb(df,lamb_no, age_range)
    
    # Actual data points for the lamb
    lamb_data = df[df['No.'] == lamb_no]
    
    # Plot the actual data points
    plt.scatter(lamb_data['Age (Days)'], lamb_data['Weight'], label=f'Lamb {lamb_no} Actual', alpha=0.6)
    
    # Plot the predicted growth curve
    plt.plot(age_range, predicted_weights, linestyle='--',color ='red', label=f'Lamb {lamb_no} Predicted')
    
    plt.title(f'Growth Curve of Lamb {lamb_no}')
    plt.xlabel('Age (Days)')
    plt.ylabel('Weight')
    plt.legend()
    plt.grid(True)
    plt.show()
```
