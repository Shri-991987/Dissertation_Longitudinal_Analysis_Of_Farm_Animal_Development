```python
df_2018test = pd.read_csv("lamb2018_test.csv")
df_2018test
```

```python
# Convert 'D.O.B' to datetime

df_2018test['D.O.B'] = df_2018test['D.O.B'].astype(str) #+ '/2016'
df_2018test['D.O.B'] = df_2018test['D.O.B'] + '/2018'


df_2018test['D.O.B'] = pd.to_datetime(df_2018test['D.O.B'], format='%d/%m/%Y')

# Extract the columns that have dates
date_columns = df_2018test.columns[2:]

# Initialize lists to store the results
lambs_list = []
age_list = []
weight_list = []

# Calculate the Age (Days) and Weight
for col in date_columns:
    age_days = (pd.to_datetime(col, format='%d/%m/%Y') - df_2018test['D.O.B']).dt.days
    lambs_list.extend(df_2018test['No.'])
    age_list.extend(age_days)
    weight_list.extend(df_2018test[col])

# Create the final DataFrame
final_df1 = pd.DataFrame({
    'No.': lambs_list,
    'Age (Days)': age_list,
    'Weight': weight_list
})

final_df1.sort_values(by=['No.', 'Age (Days)'], inplace=True)
final_df1.reset_index(drop=True, inplace=True)

print(final_df1)
```

```python
# Plotting the Curve
plt.figure(figsize=(20, 20))

for lamb_no in final_df1['No.'].unique():
    lamb_data = final_df1[final_df1['No.'] == lamb_no]
    plt.plot(lamb_data['Age (Days)'], lamb_data['Weight'], marker='o', label=f'Lamb {lamb_no}')

plt.title('Growth Curve of Lambs')
plt.xlabel('Age (Days)')
plt.ylabel('Weight')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Nadaraya–Watson Estimator on two lambs dataset
from scipy.stats import norm

def gaussian_kernel(distance, bandwidth):
    return norm.pdf(distance / bandwidth)

def nadaraya_watson(x, X, Y, bandwidth):
    weights = gaussian_kernel(x - X, bandwidth)
    return np.sum(weights * Y) / np.sum(weights)

def estimate_weight(lamb_no, target_age, bandwidth):
    lamb_data = final_df2[final_df2['No.'] == lamb_no].dropna(subset=['Weight'])
    X = lamb_data['Age (Days)'].values
    Y = lamb_data['Weight'].values
    estimated_weight = nadaraya_watson(target_age, X, Y, bandwidth)
    return estimated_weight

# Example: Estimating the weight of lamb 611 at age 200 days
lamb_no = 612
target_age = 200
bandwidth = 30  # Bandwidth Adjustment

estimated_weight = estimate_weight(lamb_no, target_age, bandwidth)
print(f"Estimated weight of lamb {lamb_no} at age {target_age} days: {estimated_weight}")
```

Using e^-c weights

```python
import numpy as np

def weight_function(age, target_age, c):
    return np.exp(-c * (age - target_age)**2)

def nadaraya_watson_normalized(target_age, X, Y, c):
    # Calculate weights using the weight function
    weights = weight_function(X, target_age, c)
    # Normalize the weights so that they sum to 1
    normalized_weights = weights / np.sum(weights)
    
    # Calculate the weighted average of Y
    estimated_weight = np.sum(normalized_weights * Y)
    
    return estimated_weight

def estimate_weight_nadaraya_watson(lamb_no, target_age, c):
    lamb_data = final_df2[final_df2['No.'] == lamb_no].dropna(subset=['Weight'])
    X = lamb_data['Age (Days)'].values
    Y = lamb_data['Weight'].values
    estimated_weight = nadaraya_watson_normalized(target_age, X, Y, c)
    return estimated_weight

# Example: Estimate the weight of lamb 611 at age 200 days
lamb_no = 612
target_age = 200
c = 0.001  # You can adjust the value of c

estimated_weight = estimate_weight_nadaraya_watson(lamb_no, target_age, c)
print(f"Estimated weight of lamb {lamb_no} at age {target_age} days: {estimated_weight}")
```

```python
# Local Linear Fit:

from sklearn.linear_model import LinearRegression

def weight_function(age, target_age, c):
    return np.exp(-c * (age - target_age)**2)


def weighted_local_linear_fit(target_age, X, Y, c):
    weights = weight_function(X, target_age, c)
    
    # Fit a weighted linear regression model
    model = LinearRegression()
    X_reshaped = X.reshape(-1, 1)
    model.fit(X_reshaped, Y, sample_weight=weights)
    
    # Predict the weight at the target age
    predicted_weight = model.predict([[target_age]])
    return predicted_weight[0]
```
